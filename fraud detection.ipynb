### Importation des modules


```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.metrics import classification_report,accuracy_score
```

### Pretraitement du dataset

```python
# ajout d'une classe appelé 'class' dans le "cas delta_otarie_IN_prepaid" > 18 donc c'est une fraude "1" (target value)
df=pd.read_csv("data_update.csv",sep=";")
df=df.drop(75)
df["delta_otarie_IN_prepaid"]=[i.replace('%', '')for i in df["delta_otarie_IN_prepaid"]]
df["delta_otarie_IN_prepaid"]=df["delta_otarie_IN_prepaid"].astype('int64')
df["class"]=[1 if i>=18 else 0 for i in df["delta_otarie_IN_prepaid"]  ]
df
```

```python
#exclure les colonnes des pourcentages et de date
df=df.drop("Delta_otarie_prs_all",axis="columns")
df=df.drop("delta_etoile6_vpn",axis="columns")
df=df.drop("day",axis="columns")
df=df.drop("delta_otarie_IN_prepaid",axis="columns")
df
```

```python
# remplacer les virgules par des points pour convertir tous les colonnes en float type
df["trafic_otarie prepaid"]=[i.replace(',', '.')for i in df["trafic_otarie prepaid"]]
df["trafic_IN prepaid"]=[i.replace(',', '.')for i in df["trafic_IN prepaid"]]
df["trafci_prs_all"]=[i.replace(',', '.')for i in df["trafci_prs_all"]]
df["trafic_otarie_all"]=[i.replace(',', '.')for i in df["trafic_otarie_all"]]
df["trafic_etoile6"]=[i.replace(',', '.')for i in df["trafic_etoile6"]]
df["trafic_etoile3"]=[i.replace(',', '.')for i in df["trafic_etoile3"]]
df["trafic_vpn"]=[i.replace(',', '.')for i in df["trafic_vpn"]]
df[["trafic_otarie prepaid","trafic_IN prepaid","trafci_prs_all","trafic_otarie_all","trafic_etoile6","trafic_etoile3","trafic_vpn"]]=df[["trafic_otarie prepaid","trafic_IN prepaid","trafci_prs_all","trafic_otarie_all","trafic_etoile6","trafic_etoile3","trafic_vpn"]].astype("float")
df
```

```python
X = df.drop(columns=['class'], axis=1)
y = df['class']
# train test split
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)
```

```python
#le oversampling pour balancer la data
from imblearn.over_sampling import SMOTE
over_sample = SMOTE()
x_smote, y_smote = over_sample.fit_resample(x_train, y_train)
```

```python
#le modele choisie
from sklearn.linear_model import LogisticRegression
m = LogisticRegression()
# training
m.fit(x_smote, y_smote)
# testing
y_pred = m.predict(x_test)
print(classification_report(y_test, y_pred))
print("F1 Score:",f1_score(y_test, y_pred))
```

```python
#enregistrement du modele
import pickle
with open('model.sav', 'wb') as f:
    pickle.dump(m, f)
```

```python
#test du modele sur des valeurs du dataset
loaded_model=pickle.load(open('model.sav','rb'))
input_data = (4225.50,3940.29,9.272366e+06,9.545026e+06,1801.833139,278.677930,12.305497	)

# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = loaded_model.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 0):
  print('non fraude')
else:
  print('fraude')
```
